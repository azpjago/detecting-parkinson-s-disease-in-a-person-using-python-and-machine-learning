**Apa itu Penyakit Parkinson?**

Penyakit Parkinson adalah gangguan progresif dari sistem saraf pusat yang mempengaruhi gerakan dan menyebabkan tremor dan kekakuan. Ini memiliki 5 tahap dan mempengaruhi lebih dari 1 juta orang setiap tahun di India. Ini kronis dan belum ada obatnya. Ini adalah gangguan neurodegeneratif yang mempengaruhi neuron penghasil dopamin di otak.

**Apa itu XGBoost?**

XGBoost adalah algoritme Pembelajaran Mesin baru yang dirancang dengan mempertimbangkan kecepatan dan kinerja. XGBoost adalah singkatan dari eXtreme Gradient Boosting dan didasarkan pada pohon keputusan. Dalam proyek ini, kami akan mengimpor XGBClassifier dari pustaka xgboost; ini adalah implementasi API scikit-learn untuk klasifikasi XGBoost.

1. Import library yang dibutuhkan

```python
import numpy as np
import pandas as pd
import os, sys
from sklearn.preprocessing import MinMaxScaler
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

2. Load dataset

```python
#DataFlair - Read the data
df=pd.read_csv('parkinsons.data')
df.head()
```

3. Dapatkan fitur dan label dari DataFrame (dataset). Fiturnya adalah semua kolom kecuali 'status', dan labelnya adalah yang ada di kolom 'status'.

```python
#DataFlair - Get the features and labels
features=df.loc[:,df.columns!='status'].values[:,1:]
labels=df.loc[:,'status'].values
```

4. Kolom 'status' memiliki nilai 0 dan 1 sebagai label; mari kita menghitung label ini untuk keduanya- 0 dan 1.

```python
#DataFlair - Get the count of each label (0 and 1) in labels
print(labels[labels==1].shape[0], labels[labels==0].shape[0])
```

5. Inisialisasi MinMaxScaler dan skalakan fitur antara -1 dan 1 untuk menormalkannya. MinMaxScaler mengubah fitur dengan menskalakannya ke rentang tertentu. Metode fit_transform() cocok dengan data dan kemudian mengubahnya. Kami tidak perlu menskalakan label.

```python
#DataFlair - Scale the features to between -1 and 1
scaler=MinMaxScaler((-1,1))
x=scaler.fit_transform(features)
y=labels
```

6. Sekarang, pisahkan dataset menjadi set pelatihan dan pengujian dengan menyimpan 20% data untuk pengujian.

```python
#DataFlair - Split the dataset
x_train,x_test,y_train,y_test=train_test_split(x, y, test_size=0.2, random_state=7)
```

7. Menginisialisasi XGBClassifier dan melatih model. Ini mengklasifikasikan menggunakan eXtreme Gradient Boosting- menggunakan algoritme peningkatan gradien untuk masalah ilmu data modern. Itu termasuk dalam kategori Ensemble Learning di ML, di mana kami melatih dan memprediksi menggunakan banyak model untuk menghasilkan satu keluaran yang unggul.

```python
#DataFlair - Train the model
model=XGBClassifier()
model.fit(x_train,y_train)
```

8. Terakhir, buat y_pred (nilai prediksi untuk x_test) dan hitung akurasi modelnya. Cetak itu.

```python
# DataFlair - Calculate the accuracy
y_pred=model.predict(x_test)
print(accuracy_score(y_test, y_pred)*100)
```
